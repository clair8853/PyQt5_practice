```python
import scanpy as sc
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import pandas as pd

adata = sc.read_h5ad('P4_marmoset_test_brain.hdf5')
```

```python
# 'SLC17A7' の発現量データフレームを取得
gene_expression_df = pd.DataFrame(adata[:, 'SLC17A7'].X.toarray(), index=adata.obs_names, columns=['SLC17A7'])
```

```python
# カスタムカラーマップを作成
cmap_green = mcolors.LinearSegmentedColormap.from_list(
    "custom_cmap", ["black", "green"]
)
```

```python
first_column = adata.obsm['spatial'][:, 0]
second_column = adata.obsm['spatial'][:, 1]
```

```python
fig, ax = plt.subplots()

fig = ax.scatter(x=first_column, 
                y=second_column, 
                c=gene_expression_df['SLC17A7'],
                cmap=cmap_green)
```
