```python
from PIL import Image, ImageOps
# 2色の画像を混ぜる

# 画像ファイルのパス
image_path1 = 'Figure_1.png'  # 画像1のパス（Redチャネルとして使用）
image_path2 = 'Figure_2.png'  # 画像2のパス（Greenチャネルとして使用）

# 画像を読み込む
img1 = Image.open(image_path1)
img2 = Image.open(image_path2)

# 画像を白黒に変換
bw_img1 = img1.convert('L')
bw_img2 = img2.convert('L')

# 画像を反転させる
inverted_img1 = ImageOps.invert(bw_img1)
inverted_img2 = ImageOps.invert(bw_img2)

# 反転させた白黒画像をそのままRGBに変換
# この時、RGBに変換してチャンネルを扱いやすくする
red_channel = inverted_img1.convert('RGB').split()[0]  # 反転させた画像をRedチャネルに
green_channel = inverted_img2.convert('RGB').split()[1]  # 反転させた画像をGreenチャネルに

# 空のBlueチャネルを作成（全てのピクセルが0の画像）
blue_channel = Image.new('L', bw_img1.size, 0)

# 新しい画像を作成する
merged_image = Image.merge('RGB', (red_channel, green_channel, blue_channel))

# 解像度を10倍に拡大
new_size = (int(merged_image.width * 10), int(merged_image.height * 10))  
high_res_image = merged_image.resize(new_size, Image.Resampling.LANCZOS)

# 高解像度で保存
high_res_image.save('merged_image.jpg', quality=100)
```

```python
from PIL import Image, ImageOps

# 이미지 파일 경로
image_path1 = 'C:/Users/visiu/Documents/pyqt5_proj/Figure_1.png'  # Red 채널로 사용할 이미지
image_path2 = 'C:/Users/visiu/Documents/pyqt5_proj/Figure_2.png'  # Green 채널로 사용할 이미지
image_path3 = 'C:/Users/visiu/Documents/pyqt5_proj/Figure_3.png'  # Blue 채널로 사용할 이미지

# 이미지 불러오기
img1 = Image.open(image_path1)
img2 = Image.open(image_path2)
img3 = Image.open(image_path3)

# 이미지 크기를 고정할 크기 (예: 500x500)
fixed_size = (500, 500)

# 이미지 크기를 고정 크기로 맞추기
img1_resized = img1.resize(fixed_size, Image.Resampling.LANCZOS)
img2_resized = img2.resize(fixed_size, Image.Resampling.LANCZOS)
img3_resized = img3.resize(fixed_size, Image.Resampling.LANCZOS)

# 이미지를 흑백으로 변환
bw_img1 = img1_resized.convert('L')
bw_img2 = img2_resized.convert('L')
bw_img3 = img3_resized.convert('L')

# 이미지를 반전
inverted_img1 = ImageOps.invert(bw_img1)
inverted_img2 = ImageOps.invert(bw_img2)
inverted_img3 = ImageOps.invert(bw_img3)

# RGB로 변환하고 각 채널 추출
red_channel = inverted_img1.convert('RGB').split()[0]  # Red 채널로 사용
green_channel = inverted_img2.convert('RGB').split()[1]  # Green 채널로 사용
blue_channel = inverted_img3.convert('RGB').split()[2]  # Blue 채널로 사용

# 새로운 이미지를 병합
merged_image = Image.merge('RGB', (red_channel, green_channel, blue_channel))

# 고해상도로 크기 조정 (원하는 배율 적용)
scale_factor = 10  # 10배로 크기 조정
new_size = (int(merged_image.width * scale_factor), int(merged_image.height * scale_factor))
high_res_image = merged_image.resize(new_size, Image.Resampling.LANCZOS)

# 결과 이미지 저장
high_res_image.show()
```
